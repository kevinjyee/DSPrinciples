## 1.1 Derive Coefficient

The cost function for Gradient Descent assuming sum of squared error is as follows:

$$J(w) = \frac{1}{2m} \sum_{i=1}^m(w^Tx^{(i)} - y^{(i)})^2$$

In SGD we only look at one sample at a time, so the cost function is:

$$J(w, (x^{(i)},y^{(i)})) = \frac{1}{2} * (w^Tx^{(i)} - y^{(i)})^2$$

Expanding out our equation to the 4 equation model stated in the problem:

$$J(w, (x^{(i)},y^{(i)})) = \frac{1}{2} * (w_0+w_1∗x_1^{(i)}+w_2∗x_1^{(i)}∗x_2^{(i)}+w_3^{(i)}∗x_2^{(i)} - y^{(i)})^2$$

At each step of our gradient secent
for $i := 1,...,m $
    $$w_j := w_j - \alpha((w_0+w_1∗x_1^{(i)}+w_2∗x_1^{(i)}∗x_2^{(i)}+w_3^{(i)}∗x_2^{(i)} - y^{(i)})x_j^{(i)}$$
and for every $ j = 0,...,n $    
This is equivalent of saying 
    $$ w_j := w_j - \alpha * \frac{\partial}{\partial w_j}J(w,(x^{(i)},y^{(i)}))$$
We take the derivative of this function for $ j = 0, 1, 2, 3 $ and update simultaneously: 

$$w_0 = w_0 - \alpha *  ((w_0+w_1∗x_1^{(i)}+w_2∗x_1^{(i)}∗x_2^{(i)}+w_3^{(i)}∗x_2^{(i)} - y^{(i)}) $$

$ $

$$w_1 = w_1 - \alpha *  ((w_0+w_1∗x_1^{(i)}+w_2∗x_1^{(i)}∗x_2^{(i)}+w_3^{(i)}∗x_2^{(i)} - y^{(i)})x_1^{(2)}$$

$ $

$$w_2 = w_2 - \alpha *  ((w_0+w_1∗x_1^{(i)}+w_2∗x_1^{(i)}∗x_2^{(i)}+w_3^{(i)}∗x_2^{(i)} - y^{(i)})x_2^{(2)}$$

$ $

$$w_3 = w_3 - \alpha *  ((w_0+w_1∗x_1^{(i)}+w_2∗x_1^{(i)}∗x_2^{(i)}+w_3^{(i)}∗x_2^{(i)} - y^{(i)})x_3^{(2)}$$

## 1.2 SGD solution the the non-linear model

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib import gridspec
from sklearn.metrics import mean_squared_error 
import sys
%matplotlib inline
```

```python
class StochasticGradientDescent:
    
    def __init__(self,learning_rate, regularization_const, num_epochs):
        '''Learning Rate, Regularization Constant, Number of Epochs'''
        self.learning_rate = learning_rate
        self.regularization_const = regularization_const
        self.num_epochs = num_epochs
    
            
    def fit(self, X, y_train, update_rule = "SGD"):
        '''X, y, Update Rule to be either SGD or Adagrad'''
        ''' Adagrad adapts learning rate to parameters performing larger updates'''
        ''' Well suited for dealing with sparse data '''
        
        
        shuffle = True
        
        if shuffle:
            shuffle_indices = np.random.permutation(np.arange(X.shape[0]))
            X = X[shuffle_indices]
            y_train = y_train[shuffle_indices]
        
        # initialize our weight matrix such that it has the 
            # same number of columns as our input feature
            W = np.array([0]*4)
            w0 = W[0]
            w1 = W[1]
            w2 = W[2]
            w3 = W[3]
            
        if(update_rule == 'SGD'):
            
            # loop over the desired number of epochs
            for epoch in range(self.num_epochs):
                # loop over our data in batches
                # For each data point, update coefficients each time
                for i in range(len(X)):
                    
                    # Initialize constants and variables at each step
                    alpha = self.learning_rate
                    reg_const = self.regularization_const
                    x1 = float(X[i][0])
                    x2 = float(X[i][1])
                    y =  float(y_train[i])
                    
                    # Update Prediction
                    yhat = w0 + (w1*x1) + (w2*x1*x2) + (w3*x2)
                    
                    # w0 does not need a regularizaiton constant
                    w0 = w0 - alpha *(yhat -y)
                    
                    # w1 to w3
                    w1 = w1 - alpha * ((yhat -y)*x1 +  reg_const*w1)
                    w2 = w2 - alpha * ((yhat - y)*x1*x2 +  reg_const*w2)
                    w3 = w3 - alpha * ((yhat - y)*x2 +  reg_const*w3)
                    
                    W = [w0, w1, w2, w3]
                #update set of weights
            self.weights = W
        
        elif(update_rule == 'adagrad'):
            ''' Adagrad modifies the general learning rate at each time step t for every paramter w'''
           
            # Adagradconst
            adaconst = 1e-6
            
            # cache of size d, number of weights
            
            Gti = np.zeros(4)
                
        
            # loop over the desired number of epochs
            for epoch in range(self.num_epochs):
                # loop over our data in batches
                # For each data point, update coefficients each time
                 for i in range(len(X)):
                    
                    # Initialize constants and variables at each step
                    alpha = self.learning_rate
                    reg_const = self.regularization_const
                    x1 = float(X[i][0])
                    x2 = float(X[i][1])
                    y =  float(y_train[i])
                    
                    # Update Prediction
                    yhat = w0 + (w1*x1) + (w2*x1*x2) + (w3*x2)
                    
                    # Update Cache for Adagrad
                    Gti[0] += (yhat-y)**2
                    Gti[1] += ((yhat -y)*x1 + reg_const*w1)**2
                    Gti[2] += ((yhat - y)*x1*x2 + reg_const*w2)**2
                    Gti[3] += ((yhat - y)*x2 + reg_const*w3)**2
                    
                    w0 = w0 - alpha *((yhat -y)/(adaconst + np.sqrt(Gti[0])))
                    
                    # w1 to w3
                    w1 = w1 - float((yhat -y)*x1 + reg_const*w1) * np.divide(alpha,(adaconst + np.sqrt(Gti[1])))
                    w2 = w2 - float((yhat - y)*x1*x2 + reg_const*w2) * np.divide(alpha,(adaconst + np.sqrt(Gti[2])))
                    w3 = w3 - float((yhat - y)*x2 + reg_const*w3) * np.divide(alpha,(adaconst + np.sqrt(Gti[3])))
                    
                    W = [w0, w1, w2, w3]
            self.weights = W
        else:
            raise ValueError('Update Rule must be either SGD or Adagrad')
            
        
        
    def predict(self, X):
        
        # Retrieve weights from fit
        weights = self.weights
        w0 = weights[0]
        w1 = weights[1]
        w2 = weights[2]
        w3 = weights[3]
            
        yhat = []
        #precit yhat
        
        for i in range(len(X)):
            x1 = float(X[i][0])
            x2 = float(X[i][1])
            yhat.append(w0 + (w1*x1) + (w2*x1*x2) + (w3*x2))
        
        return yhat
```

```python
# Read in data
samples = pd.read_csv('samples.csv')

# Split data into X and y array to pass into SGD class
X, y = np.array(samples[['x_1', 'x_2']]), np.array(samples['y'])

```

```python
alphas = [0.0001, 0.001, 0.01, 0.1, 1, 10, 100]
reg_constants = [0,10,100]
epochs = 10

MSE_const = []
Weights_const =  np.zeros((3,epochs,len(alphas),4))

for i in range(3):
    MSE_const.append(np.zeros((epochs,len(alphas))))


for constidx, const in enumerate(reg_constants):
    for alphaidx, alpha in enumerate(alphas):
        for num_epochs in range(1,epochs+1):
            sgd = StochasticGradientDescent(alpha, const, num_epochs)
            sgd.fit(X,y)
            currentMSE = MSE_const[constidx]
            currentWeight = Weights_const[constidx]
            try:
                currentMSE[num_epochs-1][alphaidx] = mean_squared_error(y,sgd.predict(X))
            except:
                currentMSE[num_epochs-1][alphaidx] = sys.maxint
            currentWeight[num_epochs-1][alphaidx] = sgd.weights
```

```python
def plot_MSE(regConstArr, index):
    # Two subplots, the axes array is 1-d
    f, axarr = plt.subplots(7, sharex=True)
    f.set_figheight(25)
    f.set_figwidth(6)
    f.tight_layout(pad = 3)
    print "Regularization Constant: " + str(regConstArr[index])

    for i in range(len(axarr)):
        axarr[i].scatter(range(epochs),np.array(MSE_const[index][:,i]))
        axarr[i].set_title("Alpha = " +str(alphas[i]) + " Regularization " + str(regConstArr[index]))
        axarr[i].set_xlabel('Epochs')
        axarr[i].set_ylabel('MSE')
```

## 1.3 Observations on Two Techniques

1. Adagrad allows for a larger initial setting of the learning rate.
It's learning rate is not fixed. Adagrad performs larger updates for infrequent and smaller updates for frequent parameters. The smaller updates for frequent parameters results in flexibility to set a larger learning rate

2. According to ruder.io, the major drawback of Adagrad is its accumulation of squared gradients in the denominator. Since every added term is positive the accumulated sum leeps growing during the training. This in turn causes the learning rate to shrink and eventually become infinitesimally small at which point the algorithm is no longer able to acquire additional knowledge.

Adadelta helps resolve the above mentioned flaw. Adadelta restricts the window of accumulated past gradients to some fixed size w.

This resolves the problem of Adagrad aggressively monotonoically decreasing it's learning rate

# Notes on Adagrad
Adagrad (http://ruder.io/optimizing-gradient-descent/) differs from vanilla SGD in that the learning rate of each weight changes over updates. A cache is maintained that holds the sum of squares of all gradients upto the current update. The learning_rate is divided by the cache, resulting in a different learning rate for each weight. A consequence of this update rule is that weights that have already seen large gradients (made large jumps) make smaller updates in subsequent iterations.
Specifically, the steps can be listed as below:
1. cache = cache + (gradients^2)
2. weights = weights + ((learning_rate)/sqrt(cache+1e-6))*gradients

# (Optional) Notes on Training with Gradient Descent
1. Compute error: This consists of a prediction error and a regularization term. From an implementation perspective, this is a function that takes as input the truth, prediction and regularization hyperparameter and returns an error
2. Compute gradients: Take a derivative of the error in terms of the weights. This can be modelled as a function that takes as input the error and features and returns the gradients for each weight
3. Update weights: Weight updates can be done using vanilla SGD or adaptive techniques. The update function takes as inputs the gradient and hyperparameters and returns the new weights.

In this question, you will explore the application of Multi-layer Perceptron (MLP) regression using sklearn package in Python. We will use the Appliances energy prediction dataset for this problem https://archive.ics.uci.edu/ml/datasets/Appliances+energy+prediction.

Following code will pre-process the data and split the data into training and test set using [train_test_split](http://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html) with **random state 42** and **test_size = 0.33**.
Make sure you have 56 variables and one target variable.

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import (train_test_split,KFold)
from sklearn.metrics import mean_squared_error
from sklearn.neural_network import MLPRegressor
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import glob
%matplotlib inline

from sklearn import datasets

data = pd.read_csv('./energydata_complete.csv') 

y = data['Appliances']
X = data.drop(['date','Appliances', 'rv1', 'rv2'], axis = 1)


from dateutil import parser
timeData = np.array(data['date'])

days = []
hours = []
for line in range(len(timeData)):
    day = parser.parse(timeData[line]).weekday()
    hour = parser.parse(timeData[line]).hour
    days.append(day)
    hours.append(hour)
    
X = pd.concat([X, pd.get_dummies(days), pd.get_dummies(hours)], axis = 1)


X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.33, random_state=42)

print(X_train.shape, y_train.shape, X_test.shape, y_test.shape)
```

## ANSWER

## 3.1 Cross Validation

```python
# Scale and center features
scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_train = pd.DataFrame(X_train)

X_test = scaler.transform(X_test)
X_test = pd.DataFrame(X_test)
```

```python
# K - Fold
cross_val = KFold(n_splits = 5, random_state = 42)

# Three Settings

mlp2 = MLPRegressor(hidden_layer_sizes=(2,), activation='tanh', solver='sgd', learning_rate='constant', random_state=42, 
                   batch_size=5000, learning_rate_init=0.005)

mlp15 = MLPRegressor(hidden_layer_sizes=(15,), activation='tanh', solver='sgd', learning_rate='constant', random_state=42, 
                   batch_size=5000, learning_rate_init=0.005)

mlp50 = MLPRegressor(hidden_layer_sizes=(50,), activation='tanh', solver='sgd', learning_rate='constant', random_state=42, 
                   batch_size=5000, learning_rate_init=0.005)

# RME for each
RMSE2 = []
RMSE15 = []
RMSE50 = []

for train, test in cross_val.split(X_train):
    mlp2.fit(X_train.iloc[train], y_train.iloc[train])
    mlp15.fit(X_train.iloc[train], y_train.iloc[train])
    mlp50.fit(X_train.iloc[train], y_train.iloc[train])
    
    RMSE2.append(np.sqrt(mean_squared_error(mlp2.predict(X_train.iloc[test]),y_train.iloc[test])))
    RMSE15.append(np.sqrt(mean_squared_error(mlp15.predict(X_train.iloc[test]),y_train.iloc[test])))
    RMSE50.append(np.sqrt(mean_squared_error(mlp50.predict(X_train.iloc[test]),y_train.iloc[test])))
    
print "Hidden Layers = 2, RMSE: " + str(np.mean(RMSE2))
print "Hidden Layers = 15, RMSE: " + str(np.mean(RMSE15))
print "Hidden Layers = 50, RMSE: " + str(np.mean(RMSE50))
```

## 3.2 RMSE on Trainig and Test Set

```python
mlp2.fit(X_train, y_train)
mlp15.fit(X_train, y_train)
mlp50.fit(X_train, y_train)

RMSE2_train = np.sqrt(mean_squared_error(mlp2.predict(X_train),y_train))
RMSE15_train = np.sqrt(mean_squared_error(mlp15.predict(X_train),y_train))
RMSE50_train = np.sqrt(mean_squared_error(mlp2.predict(X_train),y_train))

RMSE2_test = np.sqrt(mean_squared_error(mlp2.predict(X_test),y_test))
RMSE15_test = np.sqrt(mean_squared_error(mlp15.predict(X_test),y_test)) 
RMSE50_test = np.sqrt(mean_squared_error(mlp50.predict(X_test),y_test))

print('{}\nTraining Set RMSE: \n'.format('*' * 80))
print "Hidden Layers = 2, RMSE: " + str(np.mean(RMSE2_train))
print "Hidden Layers = 15, RMSE: " + str(np.mean(RMSE15_train))
print "Hidden Layers = 50, RMSE: " + str(np.mean(RMSE50_train))

print('{}\nTesting Set RMSE: \n'.format('*' * 80))
print "Hidden Layers = 2, RMSE: " + str(np.mean(RMSE2_test))
print "Hidden Layers = 15, RMSE: " + str(np.mean(RMSE15_test))
print "Hidden Layers = 50, RMSE: " + str(np.mean(RMSE50_test))
```

The 15 hidden layer model performed best on the training set.
However, the 50 hidden layer model performed the best on the testing set. 

This tells us that the addition of layers and complexity results in better accuracy on the testing set.

```python
# K - Fold
cross_val = KFold(n_splits = 5, random_state = 42)


mlp005 = MLPRegressor(hidden_layer_sizes=(50,), activation='tanh', solver='sgd', learning_rate='constant', random_state=42, 
                   batch_size=5000, learning_rate_init=0.005)
mlp01 = MLPRegressor(hidden_layer_sizes=(50,), activation='tanh', solver='sgd', learning_rate='constant', random_state=42, 
                   batch_size=5000, learning_rate_init=0.01)
mlp1 = MLPRegressor(hidden_layer_sizes=(50,), activation='tanh', solver='sgd', learning_rate='constant', random_state=42, 
                   batch_size=5000, learning_rate_init=1)
# RME for each
RMSE005 = []
RMSE01 = []
RMSE1 = []

for train, test in cross_val.split(X_train):
    mlp005.fit(X_train.iloc[train], y_train.iloc[train])
    mlp01.fit(X_train.iloc[train], y_train.iloc[train])
    mlp1.fit(X_train.iloc[train], y_train.iloc[train])
    
    RMSE005.append(np.sqrt(mean_squared_error(mlp005.predict(X_train.iloc[test]),y_train.iloc[test])))
    RMSE01.append(np.sqrt(mean_squared_error(mlp01.predict(X_train.iloc[test]),y_train.iloc[test])))
    RMSE1.append(np.sqrt(mean_squared_error(mlp1.predict(X_train.iloc[test]),y_train.iloc[test])))
    
print "Learning Rate = 0.005, RMSE: " + str(np.mean(RMSE005))
print "Learning Rate = 0.01, RMSE: " + str(np.mean(RMSE01))
print "Learning Rate = 1, RMSE: " + str(np.mean(RMSE1))
```

# Question 4: (2+5+3=10 pts) Bayes Decision Theory
a. Explain what you understand by class-conditional likelihood, class priors, and posterior probability of a class given an input, and the relationship between them. Please define all symbols and equations used explicitly.
( 2 points )


b. Suppose you want to learn a binary classifier to predict whether or not a customer will buy a TV. The class label is 0 if the customer does not buy a TV and 1 if they do. For each customer, you are given two features, $x_1$ is the per hour salary and $x_2$ is the age. Assume that the class conditional distribution $p([x_1 , x_2]|C)$ is Gaussian. The mean salary and age of the people who do buy a TV is 15 and 30 respectively and that of those who don’t is 8.5 and 25. Assume that both classes of customers have the same covariance given by the Identity matrix. Further, your sales data suggests that only 1 in 3 people actually bought a TV in the last few years. Mathematically derive the (optimal) Bayes decision boundary for this problem. (5 points)

c. Now sample 100 customers from each class (C = 0, 1) under the assumed distribution and the estimated parameters and plot their features. Additionally, plot the decision boundary you obtained in the part (b) on the same plot. (3 points)

## ANSWER

### 4.a 

**Class Conditional likelihood function:** $p(x|C_i)$, The probability density function for feature $x$ given that the state of the nature is labeled $C_i$. This relates the probability that a feature can take on a certain value given the class.

**Class-Priors:** $P(C = C_i)$, The prior probabiliy reflects our knowledge of how likely we expect a certain classification to occur before we can actually observe said state of nature.

**Posterior Probability:** $P(C_i|x)$, The posterior probability is the probability that an object is in a classificiation given our observation of feature x.

**Relationship - Bayes Formula to make decision** 

$$posterior~probability = \frac{conditional~probability \dot{} prior~probability}{evidence}$$

$$p(C_i|x_j) = \frac{p(x_j|C_i) \dot{} p(C_i)}{p(x_j)}$$

### 4.b
Let $C_0$ represent the class label if the customer does not buy a TV and $C_1$ if the customer does.
And let's have features $x_1$ denote their salary per hour and $x_2$ denote their age.

The optimial bayes decision boundary is given that:

$$P(C_0|(x_1,x_2)) = P(C_1|(x_1,x_2))$$

Using Bayes rule, this can be represented as: 

$$P(C_0)P((x_1,x_2)|C_0) = P(C_1)P((x_1,x_2)|C_1)$$

Using the gaussian model we can interpret this as:

$$\frac{2}{3} \frac{1}{\sqrt{2 \pi}} e^{(\frac{-[(x_1-8.5)^2 + (x_2 - 25)^2}{2}])} = \frac{1}{3} \frac{1}{\sqrt{2 \pi}} e^{(\frac{-[(x_1-15)^2 + (x_2 - 30)^2}{2}])} $$

Simplify:

$$ ln2 - \frac{1}{2} [(x_1-8.5)^2 + (x_2 - 25)^2] = - \frac{1}{2}[(x_1-8.5)^2 + (x_2 - 25)^2]$$

$ $

$$x_2 = \frac{-13 (x_1 -33)}{10}$$

The decision boundary is given by $x_2 = \frac{-13 (x_1 -33)}{10}$

### 4.c

```python
import numpy as np
import seaborn as sns
import math
import matplotlib.pyplot as plt 

mean_0 = [8.5, 25]
cov_0 = [[1,0], [0, 1]]
mean_1 = [15, 30]
cov_1 = [[1, 0], [0, 1]]

c0 = np.random.multivariate_normal(mean_0, cov_0, 100)
c1 = np.random.multivariate_normal(mean_1, cov_1, 100)
x1 = np.linspace(4,20, num= 1000)
x2 = (-float(13)/10) * (x1-33)
```

```python
plt.scatter(c0[:,0], c0[:,1], label='Customers that do not buy TVs')
plt.scatter(c1[:,0], c1[:,1], label='Customers that buy TVs')
plt.plot(x1, x2, 'r--')
plt.xlabel('Salary')
plt.ylabel('Age')
plt.title('Decision Boundary')
plt.legend()
plt.show()
```

# Question 5: Decision Tree using Python (10 pts)
In this problem, you will model the data using decision trees to perform a classification task. Load the BreastTissue.csv dataset in the Q1 folder. The dataset has been preprocessed. The description of the dataset can be found [here](https://archive.ics.uci.edu/ml/datasets/Breast+Tissue). Using the class tree.DecisionTreeClassifier (http://scikit-learn.org/stable/modules/classes.html#module-sklearn.tree), build two different trees with a maximum depth of two using the split criteria (i) Gini and (ii) Entropy. Use entire data to build trees. Plot the two trees. If your classifier object is called clf, use the following commands to save the generated tree as a '.dot' file that can be used to visualize the tree using Webgraphviz: http://www.webgraphviz.com/

Hint: see  http://scikit-learn.org/stable/modules/generated/sklearn.tree.export_graphviz.html

Copy paste the contents of your '.dot' file into the text window on the website to visualize the trees. At which node(s) do they differ?
## ANSWER

```python
from sklearn import tree
from sklearn.externals.six import StringIO
import numpy as np
import pandas as pd

data = pd.read_csv("BreastTissue.csv")

X = data.drop(['Case #', 'Class'], axis=1)
y = data.Class

gini_clf = tree.DecisionTreeClassifier(criterion='gini', max_depth = 2)
gini_clf.fit(X,y)

with open("decision_tree_gini.dot", 'w') as f:
    f = tree.export_graphviz(gini_clf, out_file=f)
    
entropy_clf = tree.DecisionTreeClassifier(criterion='entropy', max_depth = 2)
entropy_clf.fit(X,y)

with open("decision_tree_entropy.dot", 'w') as f:
    f = tree.export_graphviz(entropy_clf, out_file=f)
```
