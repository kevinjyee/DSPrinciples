## 1.1 Derive Coefficient

The cost function for Gradient Descent assuming sum of squared error is as follows:

$$J(w) = \frac{1}{2m} \sum_{i=1}^m(w^Tx^{(i)} - y^{(i)})^2$$

In SGD we only look at one sample at a time, so the cost function is:

$$J(w, (x^{(i)},y^{(i)})) = \frac{1}{2} * (w^Tx^{(i)} - y^{(i)})^2$$

Expanding out our equation to the 4 equation model stated in the problem:

$$J(w, (x^{(i)},y^{(i)})) = \frac{1}{2} * (w_0+w_1∗x_1^{(i)}+w_2∗x_1^{(i)}∗x_2^{(i)}+w_3^{(i)}∗x_2^{(i)} - y^{(i)})^2$$

At each step of our gradient secent
for $i := 1,...,m $
    $$w_j := w_j - \alpha((w_0+w_1∗x_1^{(i)}+w_2∗x_1^{(i)}∗x_2^{(i)}+w_3^{(i)}∗x_2^{(i)} - y^{(i)})x_j^{(i)}$$
and for every $ j = 0,...,n $    
This is equivalent of saying 
    $$ w_j := w_j - \alpha * \frac{\partial}{\partial w_j}J(w,(x^{(i)},y^{(i)}))$$
We take the derivative of this function for $ j = 0, 1, 2, 3 $ and update simultaneously: 

$$w_0 = w_0 - \alpha *  ((w_0+w_1∗x_1^{(i)}+w_2∗x_1^{(i)}∗x_2^{(i)}+w_3^{(i)}∗x_2^{(i)} - y^{(i)}) $$

$ $

$$w_1 = w_1 - \alpha *  ((w_0+w_1∗x_1^{(i)}+w_2∗x_1^{(i)}∗x_2^{(i)}+w_3^{(i)}∗x_2^{(i)} - y^{(i)})x_1^{(2)}$$

$ $

$$w_2 = w_2 - \alpha *  ((w_0+w_1∗x_1^{(i)}+w_2∗x_1^{(i)}∗x_2^{(i)}+w_3^{(i)}∗x_2^{(i)} - y^{(i)})x_2^{(2)}$$

$ $

$$w_3 = w_3 - \alpha *  ((w_0+w_1∗x_1^{(i)}+w_2∗x_1^{(i)}∗x_2^{(i)}+w_3^{(i)}∗x_2^{(i)} - y^{(i)})x_3^{(2)}$$

## 1.2 SGD solution the the non-linear model

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib import gridspec
from sklearn.metrics import mean_squared_error 
import sys
%matplotlib inline
```

```python
class StochasticGradientDescent:
    
    def __init__(self,learning_rate, regularization_const, num_epochs):
        '''Learning Rate, Regularization Constant, Number of Epochs'''
        self.learning_rate = learning_rate
        self.regularization_const = regularization_const
        self.num_epochs = num_epochs
    
            
    def fit(self, X, y_train, update_rule = "SGD"):
        '''X, y, Update Rule to be either SGD or Adagrad'''
        ''' Adagrad adapts learning rate to parameters performing larger updates'''
        ''' Well suited for dealing with sparse data '''
        
        
        shuffle = True
        
        if shuffle:
            shuffle_indices = np.random.permutation(np.arange(X.shape[0]))
            X = X[shuffle_indices]
            y_train = y_train[shuffle_indices]
        
        # initialize our weight matrix such that it has the 
            # same number of columns as our input feature
            W = np.array([0]*4)
            w0 = W[0]
            w1 = W[1]
            w2 = W[2]
            w3 = W[3]
            
        if(update_rule == 'SGD'):
            
            # loop over the desired number of epochs
            for epoch in range(self.num_epochs):
                # loop over our data in batches
                # For each data point, update coefficients each time
                for i in range(len(X)):
                    
                    # Initialize constants and variables at each step
                    alpha = self.learning_rate
                    reg_const = self.regularization_const
                    x1 = float(X[i][0])
                    x2 = float(X[i][1])
                    y =  float(y_train[i])
                    
                    # Update Prediction
                    yhat = w0 + (w1*x1) + (w2*x1*x2) + (w3*x2)
                    
                    # w0 does not need a regularizaiton constant
                    w0 = w0 - alpha *(yhat -y)
                    
                    # w1 to w3
                    w1 = w1 - alpha * ((yhat -y)*x1 + 2 *reg_const*w1)
                    w2 = w2 - alpha * ((yhat - y)*x1*x2 + 2* reg_const*w2)
                    w3 = w3 - alpha * ((yhat - y)*x2 + 2 * reg_const*w3)
                    
                    W = [w0, w1, w2, w3]
                #update set of weights
            self.weights = W
        
        elif(update_rule == 'adagrad'):
            ''' Adagrad modifies the general learning rate at each time step t for every paramter w'''
           
            # Adagradconst
            adaconst = 1e-6
            
            # cache of size d, number of weights
            
            Gti = np.zeros(4)
                
        
            # loop over the desired number of epochs
            for epoch in range(self.num_epochs):
                # loop over our data in batches
                # For each data point, update coefficients each time
                 for i in range(len(X)):
                    
                    # Initialize constants and variables at each step
                    alpha = self.learning_rate
                    reg_const = self.regularization_const
                    x1 = float(X[i][0])
                    x2 = float(X[i][1])
                    y =  float(y_train[i])
                    
                    # Update Prediction
                    yhat = w0 + (w1*x1) + (w2*x1*x2) + (w3*x2)
                    
                    # Update Cache for Adagrad
                    Gti[0] += (yhat-y)**2
                    Gti[1] += ((yhat -y)*x1 + reg_const*w1)**2
                    Gti[2] += ((yhat - y)*x1*x2 + reg_const*w2)**2
                    Gti[3] += ((yhat - y)*x2 + reg_const*w3)**2
                    
                    w0 = w0 - alpha *((yhat -y)/(adaconst + np.sqrt(Gti[0])))
                    
                    # w1 to w3
                    w1 = w1 - float((yhat -y)*x1 + reg_const*w1) * np.divide(alpha,(adaconst + np.sqrt(Gti[1])))
                    w2 = w2 - float((yhat - y)*x1*x2 + reg_const*w2) * np.divide(alpha,(adaconst + np.sqrt(Gti[2])))
                    w3 = w3 - float((yhat - y)*x2 + reg_const*w3) * np.divide(alpha,(adaconst + np.sqrt(Gti[3])))
                    
                    W = [w0, w1, w2, w3]
            self.weights = W
        else:
            raise ValueError('Update Rule must be either SGD or Adagrad')
            
        
        
    def predict(self, X):
        
        # Retrieve weights from fit
        weights = self.weights
        w0 = weights[0]
        w1 = weights[1]
        w2 = weights[2]
        w3 = weights[3]
            
        yhat = []
        #precit yhat
        
        for i in range(len(X)):
            x1 = float(X[i][0])
            x2 = float(X[i][1])
            yhat.append(w0 + (w1*x1) + (w2*x1*x2) + (w3*x2))
        
        return yhat
```

```python
# Read in data
samples = pd.read_csv('samples.csv')

# Split data into X and y array to pass into SGD class
X, y = np.array(samples[['x_1', 'x_2']]), np.array(samples['y'])

```

```python
alphas = [0.0001, 0.001, 0.01, 0.1, 1, 10, 100]
reg_constants = [0,10,100]
epochs = 10

MSE_const = []
Weights_const =  np.zeros((3,epochs,len(alphas),4))

for i in range(3):
    MSE_const.append(np.zeros((epochs,len(alphas))))


for constidx, const in enumerate(reg_constants):
    for alphaidx, alpha in enumerate(alphas):
        for num_epochs in range(1,epochs+1):
            sgd = StochasticGradientDescent(alpha, const, num_epochs)
            sgd.fit(X,y)
            currentMSE = MSE_const[constidx]
            currentWeight = Weights_const[constidx]
            try:
                currentMSE[num_epochs-1][alphaidx] = mean_squared_error(y,sgd.predict(X))
            except:
                currentMSE[num_epochs-1][alphaidx] = sys.maxint
            currentWeight[num_epochs-1][alphaidx] = sgd.weights
```

```python
def plot_MSE(regConstArr, index):
    # Two subplots, the axes array is 1-d
    f, axarr = plt.subplots(7, sharex=True)
    f.set_figheight(25)
    f.set_figwidth(6)
    f.tight_layout(pad = 3)
    print "Regularization Constant: " + str(regConstArr[index])

    for i in range(len(axarr)):
        axarr[i].scatter(range(epochs),np.array(MSE_const[index][:,i]))
        axarr[i].set_title("Alpha = " +str(alphas[i]) + " Regularization " + str(regConstArr[index]))
        axarr[i].set_xlabel('Epochs')
        axarr[i].set_ylabel('MSE')
```

# Notes on Adagrad
Adagrad (http://ruder.io/optimizing-gradient-descent/) differs from vanilla SGD in that the learning rate of each weight changes over updates. A cache is maintained that holds the sum of squares of all gradients upto the current update. The learning_rate is divided by the cache, resulting in a different learning rate for each weight. A consequence of this update rule is that weights that have already seen large gradients (made large jumps) make smaller updates in subsequent iterations.
Specifically, the steps can be listed as below:
1. cache = cache + (gradients^2)
2. weights = weights + ((learning_rate)/sqrt(cache+1e-6))*gradients

# (Optional) Notes on Training with Gradient Descent
1. Compute error: This consists of a prediction error and a regularization term. From an implementation perspective, this is a function that takes as input the truth, prediction and regularization hyperparameter and returns an error
2. Compute gradients: Take a derivative of the error in terms of the weights. This can be modelled as a function that takes as input the error and features and returns the gradients for each weight
3. Update weights: Weight updates can be done using vanilla SGD or adaptive techniques. The update function takes as inputs the gradient and hyperparameters and returns the new weights.
